## Génération de chemins aléatoires

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_paths_mono(nb_paths, nb_step, sigma, r, spot, T):
    # Pas de temps
    dt = T / nb_step
    
    # Calcul en une fois des parties drift et diffusion
    W=np.random.standard_normal(size=(nb_paths, nb_step))
    
    exp_diffus_array = np.exp(sigma*np.sqrt(dt)*W)
    exp_drift = np.exp((r-sigma**2/2)*dt)
    
    # Initialisation de la matrice contenant les chemins
    paths_array = np.zeros((nb_paths, nb_step+1), np.float64)
    paths_array[:, 0] = spot # Valeur de la première colonne
    
    # Calcul des chemins
    for i in range(nb_paths):
        for j in range(nb_step):
            paths_array[i,j+1] = paths_array[i,j]*exp_drift*exp_diffus_array[i, j]
    
    return paths_array
```

```python
# Paramètres
nb_paths=100
nb_step=252
sigma=3e-2
r=1e-2
spot=100.
T=1.

# Génération des chemins
np.random.seed(123)
paths_array = generate_paths_mono(nb_paths, nb_step, sigma, r, spot, T)

# Tracés des chemins
X=[T/nb_step*i for i in range(nb_step+1)]
for traj_i in range(nb_paths):
    plt.plot(X, paths_array[traj_i])
plt.show()
```

## Pricing d'un call/put vanille (BS)

### Expression analytique

On commence par définir les fonctions renvoyant les prix par la formule analytique de Black-Scholes.

```python
from scipy.stats import norm

def d1_funct(spot, strike, r, sigma, T):
    return (np.log(spot/strike) + (r + sigma**2/2)*T) / (sigma * np.sqrt(T))
    
def d2_funct(d1, sigma, T):
    return d1 - sigma*np.sqrt(T)

def Call_BS(strike, spot, r, T, sigma):
    d1 = d1_funct(spot, strike, r, sigma, T)
    d2 = d2_funct(d1, sigma, T)
    
    return spot*norm.cdf(d1) - strike*np.exp(-r*T)*norm.cdf(d2)

def Put_BS(strike, spot, r, T, sigma):
    d1 = d1_funct(spot, strike, r, sigma, T)
    d2 = d2_funct(d1, sigma, T)
    
    return strike*np.exp(-r*T)*norm.cdf(-d2) - spot*norm.cdf(-d1)
```

### Calcul par Monte-Carlo

On définit les fonctions qui permettent de calculer le prix d'un Call/Put par Monte-Carlo, avec une dynamique de sous-jacent log-normal, sans technique de réduction de variance.

```python
def Call_BS_MC(n, strike, spot, r, T, sigma):
    discount_factor = np.exp(-r*T)
    
    # Calcul de n valeurs du sous-jacent à la maturité T, en une fois
    X = np.random.standard_normal(size=n)
    S = spot*np.exp((r-sigma**2/2)*T + sigma*np.sqrt(T)*X)

    # Calcul du payoff des n simulations, de la moyenne et de l'écart-type MC des payoff
    payoff_array = np.where(S > strike, S - strike, 0) # Retourne un array qui vérifie (S_T - K)+
    mean_payoffactu, std_payoffactu = np.mean(discount_factor*payoff_array), np.std(discount_factor*payoff_array)
    
    return (mean_payoffactu, std_payoffactu)

def Put_BS_MC(n, strike, spot, r, T, sigma):
    discount_factor = np.exp(-r*T)
    
    # Calcul de la valeur à maturité du sous-jacent en une fois
    X = np.random.standard_normal(size=n)
    S = spot*np.exp((r-sigma**2/2)*T + sigma*np.sqrt(T)*X)

    # Calcul du payoff des n simulations, de la moyenne et de l'écart-type des payoff MC
    payoff_array = np.where(strike > S, strike - S, 0) # Retourne un array qui vérifie (K - S_T)+
    mean_payoffactu, std_payoffactu = np.mean(discount_factor*payoff_array), np.std(discount_factor*payoff_array)
    
    return (mean_payoffactu, std_payoffactu)
```

```python
n=1000
sigma=2e-2
r=1e-4
spot=100.
strike=102
T=1.

np.random.seed(123)
price_MC, std_MC = Call_BS_MC(n, strike, spot, r, T, sigma)

price_BS = Call_BS(strike, spot, r, T, sigma)

print("Prix obtenu par MC : {} (+- {})".format(price_MC, 1.96*std_MC/np.sqrt(n))) # Seuil 5%
print("Prix théorique sous BS: {}".format(price_BS))
```

```python
n=1000
sigma=2e-2
r=1e-4
spot=100.
strike=98
T=1.

np.random.seed(123)
price_MC, std_MC = Put_BS_MC(n, strike, spot, r, T, sigma)[0:2]

price_BS = Put_BS(strike, spot, r, T, sigma)

print("Prix obtenu par MC : {} (+- {})".format(price_MC, 1.96*std_MC/np.sqrt(n))) # Seuil 5%
print("Prix théorique sous BS: {}".format(price_BS))
```

### Calcul par Monte-Carlo avec réduction de variance

On définit désormais la fonction prix d'un call par Monte-Carlo avec variable de contrôle afin de réduire la variance Monte-Carlo (pour $\frac{K}{S_0}$ pas trop grand par rapport à $1$).

Par la parité call/put, $C_{t=0} = P_{t=0} + S_0 - Ke^{-rT}$

```python
def Call_BS_MC_v2(n, strike, spot, r, T, sigma):
    discount_factor = np.exp(-r*T)

    mean_put, std_put = Put_BS_MC(n, strike, spot, r, T, sigma)

    return (mean_put + spot - strike*discount_factor, std_put)
```

```python
n=10000
sigma=0.3
r=0.05
spot=100.
strike=105
T=1.

np.random.seed(123)
price_MC, std_MC = Call_BS_MC_v2(n, strike, spot, r, T, sigma)
price_MC_v1, std_MC_v1 = Call_BS_MC(n, strike, spot, r, T, sigma)

price_BS = Call_BS(strike, spot, r, T, sigma)

print("Prix obtenu par MC avec réduction de variance : {} (+- {})"
      .format(price_MC, 1.96*std_MC/np.sqrt(n))) # Proba à 95%

print("Prix obtenu par MC : {} (+- {})".format(price_MC_v1, 1.96*std_MC_v1/np.sqrt(n))) # Proba à 95%

print("Prix théorique sous BS: {}".format(price_BS))

```

## Pricing d'un autocall

### Autocall sans put

On considère un autocall avec les caractéristiques suivantes (inspiré du livre de M. Bouzoubaa, <i>Exotic Options and Hybrids </i>) :

- Durée de vie maximale : $T$

- Un sous-jacent $S$

- $n$ dates d'observation du sous-jacent : $0 < t_1 < ... < t_n = T$

- Octroie un coupon $N.C$ ($N$ : nominal, $C$ : pourcent de nominal) au temps $t_i$ si:
    - le sous-jacent est au-dessus du seuil $K$ à la date d'observation $t_i$
    - <b> et </b> le niveau du sous-jacent n'a jamais été observé au-dessus d'un seuil $B$ aux temps $t_1, \dots, t_{i-1}$.

- Rembourse le notionnel au temps $t_i$ et désactive le produit si:
    - la valeur du sous-jacent en $t_i$ est observé au-dessus de $B$
    - <b> ou </b> le produit atteint sa durée de vie maximale.

$Coupon(t_i) = NC\mathbb{1}_{\{S_{t_i} > K\}} \mathbb{1}_{\{max_{k = 1, ..., i-1}S_{t_k} < B\}}$

$Rebate(t_i) = N(\mathbb{1}_{\{S_{t_i} > B\}} \mathbb{1}_{\{max_{k = 1, ..., i-1}S_{t_k} < B\}} + \mathbb{1}_{\{max_{k = 1, ..., n} S_{t_k} < B\}})$

#### Pricer naïf

On commence par écrire un pricer non optimisé utilisant des boucles Python, puis on optimisera le pricer dans la partie suivante en utilisant les fonctions NumPy.

```python
def autocall_mono(K, B, C, N, T, spot, r, sigma, nb_observ, nb_paths):
    """
        Pricer d'autocall par Monte-Carlo sous Black-Scholes, non-optimisé
        ----------
        Paramètres :
        
        K : Strike
        B : Seuil de désactivation du produit
        C : Coupon (en % de N)
        N : Nominal remboursé
        T : Durée de vie maximale de l'autocall (année(s))
        spot : Valeur spot du sous-jacent
        r : taux d'intérêt sans risque
        sigma : volatilité du sous-jacent
        nb_observ : nombre total d'observations du prix
        nb_paths : nombre de trajectoires à simuler
        
        Retour :
        Couple (a,b), où a : moyenne des payoffs actu
                         b : écart-type des payoffs actu
        ----------
    """
    
    dt = T/nb_observ # pas de temps
    total_list = [] # liste des payoff pour toutes les trajectoires
    paths_array = generate_paths_mono(nb_paths, nb_observ, sigma, r, spot, T) # matrice de trajectoires (une par ligne)
    
    # Boucle sur les trajectoires
    for i in range(nb_paths):
        j = 1
        t = dt
        total = 0
        
        # Boucle sur le chemin d'une trajectoire
        while True:
            S = paths_array[i, j]
            
            # Détachement d'un coupon si S est au-dessus du strike
            if (S > K): # and S < B:
                total += N*C*np.exp(-r*t)
            
            # Désactivation du produit si S est au-dessus de la barrière désactivante ou si le produit atteint sa durée de vie max
            if (S > B) or (j == nb_observ):
                total += N*np.exp(-r*t)
                total_list.append(total)
                break
                 
            t = t + dt
            j = j + 1
    
    return np.mean(total_list), np.std(total_list)
```

```python
# Paramètres des chemins
nb_paths=1000
nb_observ, T = 16, 4. # 16 observations en 4 ans i.e. 1 observation tous les trois mois
sigma=8e-2
r=2e-2
spot=100.

# Strike et barrière désactivante de l'autocall
K = 80
B = 100

# Génération des chemins
np.random.seed(123)
paths_array = generate_paths_mono(nb_paths, nb_observ, sigma, r, spot, T)

# Tracés des chemins
X=[T/nb_observ*i for i in range(nb_observ+1)]
for traj_i in range(nb_paths):
    plt.plot(X, paths_array[traj_i])
    
plt.plot(X, [K for x in range(len(X))], label="Strike")
plt.plot(X, [B for x in range(len(X))], label="Barrière désactivante")
plt.legend(loc="upper left")
plt.show()
```

```python
N = 1 # Nominal remboursé
C = 8e-2 # Coupon

nb_paths=10000
nb_observ, T = 16, 4. # 16 observations en 4 ans i.e. 1 observation tous les trois mois
sigma=8e-2
r=2e-2
spot=100.

# Strike et barrière désactivante de l'autocall
K = 80
B = 100
```

```python
np.random.seed(123)
price_MC, std_MC = autocall_mono(K, B, C, N, T, spot, r, sigma, nb_observ, nb_paths)
print("Prix de l'autocall : {} (+- {})".format(price_MC, 1.96*std_MC/np.sqrt(nb_paths)))
```

#### Pricer optimisé

Une grande partie des fonctions NumPy sont écrites en C, ce qui permet d'accélerer les calculs d'espérance Monte-Carlo.

```python
# Optimisation du pricer en utilisant les opérations NumPy
def autocall_mono_opti(K, B, C, N, T, spot, r, sigma, nb_observ, nb_paths):
    """
        Pricer d'autocall par Monte-Carlo sous Black-Scholes, optimisé à l'aide des fonctions NumPy
        ----------
        Paramètres :
        
        K : Strike
        B : Seuil de désactivation du produit
        C : Coupon (en % de N)
        N : Nominal remboursé
        T : Durée de vie maximale de l'autocall (année(s))
        spot : Valeur spot du sous-jacent
        r : taux d'intérêt sans risque
        sigma : volatilité du sous-jacent
        nb_observ : nombre total d'observations du prix
        nb_paths : nombre de trajectoires à simuler
        
        Retour :
        Couple (a,b), où a : moyenne des payoffs actu
                         b : écart-type des payoffs actu
        ----------
    """
    
    dt = T/nb_observ # pas de temps
    paths_array = generate_paths_mono(nb_paths, nb_observ, sigma, r, spot, T) # matrice de trajectoires (une par ligne)
    
    # indices (pour chaques trajectoires) à partir desquels l'autocall est désactivé
    index_deactiv_array = np.argmax(paths_array > B, axis = 1)
    index_deactiv_array = np.where(index_deactiv_array == 0, nb_observ, index_deactiv_array) # car si le produit atteint sa durée de vie max alors NumPy donne indice de désactivation égal à 0

    
    # Calcul du remboursement du notionnel par trajectoire
    #rebate_array = np.where(index_deactiv_array == 0, N*np.exp(-r*T), N*np.exp(-r*dt*index_deactiv_array))
    rebate_array = N*np.exp(-r*dt*index_deactiv_array)
    
    # Calcul des coupons payés (une ligne par trajectoire)
    coupons_actu_array = N*C*np.exp(-r*dt*np.arange(nb_observ + 1)) # Valeur des coupons actualisées pour chaques temps d'obs
    coupons_actu_array[0] = 0 # Aucun coupon n'est versé à la date initiale
    
    coupons_striked_array = np.where(paths_array > K, coupons_actu_array, 0) # vaut N.C actualisé si S > K
    coupons_cumsum_array = np.cumsum(coupons_striked_array, axis = 1) # somme cumulée des coupons, par trajectoire
    
    coupons_paid_array = coupons_cumsum_array[range(nb_paths), index_deactiv_array] # Somme des coupons actu jusqu'à désactivation du produit
    
    # Payoff par trajectoire
    payoff_array = rebate_array + coupons_paid_array # Remboursement du notionnel (actualisé) + coupons actualisés
    
    return np.mean(payoff_array), np.std(payoff_array)
```

```python
from timeit import default_timer as timer

N = 1 # Nominal remboursé
C = 10e-2 # Coupon

nb_paths=100000
nb_observ, T = 12, 3. # 12 observations en 3 ans i.e. 1 observation tous les trois mois
sigma=2e-2
r=1e-2
spot=100.

# Strike et barrière désactivante de l'autocall
K = 100
B = 110
```

```python
# Non optimisé
np.random.seed(123)

start = timer()
price_MC, std_MC = autocall_mono(K, B, C, N, T, spot, r, sigma, nb_observ, nb_paths)
end = timer()

print("Prix de l'autocall : {} (+- {})".format(price_MC, 1.96*std_MC/np.sqrt(nb_paths)))
print("Calculé en {}s sans NumPy".format(end - start), end="\n\n")

# Optimisé
np.random.seed(123)

start = timer()
price_MC, std_MC = autocall_mono_opti(K, B, C, N, T, spot, r, sigma, nb_observ, nb_paths)
end = timer()

print("Prix de l'autocall : {} (+- {})".format(price_MC, 1.96*std_MC/np.sqrt(nb_paths)))
print("Calculé en {}s avec NumPy".format(end - start))
```

### Autocall avec put

On considère un produit composé:
- d'un long autocall
- et d'un short put européen down-and-in avec observations journalières de la barrière

Caractéristiques de l'autocall:
- Durée de vie maximale : $T$

- Un sous-jacent $S$

- $n$ dates d'observation du sous-jacent : $0 < t_1 < ... < t_n = T$

- Octroie un coupon $N.C$ ($N$ : nominal, $C$ : pourcent de nominal) au temps $t_i$ si:
    - le sous-jacent est au-dessus du seuil $K$ à la date d'observation $t_i$
    - <b> et </b> le niveau du sous-jacent n'a jamais été observé au-dessus d'un seuil $B$ aux temps $t_1, \dots, t_{i-1}$.

- Rembourse le notionnel au temps $t_i$ et désactive le produit si:
    - la valeur du sous-jacent en $t_i$ est observé au-dessus de $B$
    - <b> ou </b> le produit atteint sa durée de vie maximale.

Caractéristiques du put:

- Put down-and-in européen de maturité $T$
- Prix d'exercice $K_{PDI}$
- Barrière down-and-in $B_{PDI}$ regular ($K_{PDI} < B_{PDI}$) observée quotidiennement
- Nominal : $N_{PDI}$

```python
# To be tested and optimized
def autocall_mono_PDI(K, B, C, N, T, spot, r, sigma, K_pdi, B_pdi, N_pdi, nb_observ, nb_paths):
    """
        Pricer d'autocall avec put down-and-in par Monte-Carlo sous Black-Scholes
        ----------
        Paramètres :
        
        K : Strike
        B : Seuil de désactivation du produit
        C : Coupon (en % de N)
        N : Nominal remboursé
        T : Durée de vie maximale de l'autocall (année(s)), ENTIER
        spot : Valeur spot du sous-jacent
        r : taux d'intérêt sans risque
        sigma : volatilité du sous-jacent
        nb_observ : nombre d'observations de prix PAR AN
        nb_paths : nombre de trajectoires à simuler
        
        Retour :
        Couple (a,b), où a : moyenne des payoffs actu
                         b : écart-type des payoffs actu
        ----------
    """
    
    # Fonctions non implémentées (barriere non régulière, T non entier)
    if (B_pdi < K_pdi) or (np.round(T) != T):
        raise NotImplemented
        
    dt = T/nb_observ # pas de temps
    total_list = [] # liste des payoff pour toutes les trajectoires
    
    # On génère une matrice de trajectoires (une trajectoire par ligne) pour une observation quotidienne du sous-jacent
    paths_array = generate_paths_mono(nb_paths, int(np.round(252*T)), sigma, r, spot, T) 
    
    # Boucle sur les trajectoires
    for i in range(nb_paths):
        j = 1
        t = dt
        pdi_active_bool = False # La barrière n'est pas observée à la date t=0
        autocall_breached_bool = False
        total = 0
        
        # Boucle sur le chemin d'une trajectoire
        while True:
            S = paths_array[i, j]
            
            # Activation du put si le sous-jacent passe en dessous de la barrière PDI
            if (S < B_pdi):
                pdi_active_bool = True
            
            # Détachement d'un coupon si S au-dessus du strike
            if (S > K):# and (S < B):
                total += N*C*np.exp(-r*t)
            
            # Désactivation du produit si S au-dessus de la barrière désactivante de l'autocall
            if (S > B):
                total += N*np.exp(-r*t)
                autocall_breached_bool = True
            
            # A la dernière date d'observation
            if (j == nb_observ):
                
                # Si le produit a atteint sa durée de vie max sans dépasser la barrière, on rembourse le notionnel
                if (autocall_breached_bool == False):
                    total += N*np.exp(-r*t)
                    
                # Calcul et retranchement de la valeur du put à maturité, s'il a été activé
                if pdi_active_bool:
                    total -= np.max(K_pdi - S, 0)
                
                total_list.append(total)
                break
                 
            t = t + dt
            j = j + 1
    
    return np.mean(total_list), np.std(total_list)
```

```python
np.random.seed(123)
autocall_mono_opti(K, B, C, N, T, spot, r, sigma, nb_observ, nb_paths)
```

```python
np.random.seed(123)
autocall_mono_PDI(K, B, C, N, T, spot, r, sigma, K_pdi, B_pdi, N_pdi, nb_observ, nb_paths)
```

Le prix de l'autocall ainsi calculé est de 97% < 100%.
